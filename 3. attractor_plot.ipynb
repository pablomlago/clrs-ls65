```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4))

x = np.arange(6, 14)
y = np.array([[0.9497, 0.9456, 0.9387, 0.9329, 0.9241, 0.9171, 0.9073, 0.8955],
              [0.9545, 0.9490, 0.9403, 0.9318, 0.9226, 0.9110, 0.8955, 0.8909],
              [0.9556, 0.9480, 0.9413, 0.9319, 0.9226, 0.9120, 0.8968, 0.8815]])
# ax1.plot(x,y, '--')
ax1.set_xticks(np.arange(6,14))
ax1.scatter(x, y.mean(axis=0), marker='o', s=20)
ax1.errorbar(x, y.mean(axis=0), yerr=y.std(axis=0), linestyle='--', capsize=4)
ax1.set_xlabel('Step Count')
ax1.set_ylabel('Accuracy')

x = np.arange(-3, 5)
y = np.array([[0.8198, 0.8949, 0.9308, 0.9368, 0.9286, 0.9200, 0.9147, 0.9129],
              [0.8219, 0.8961, 0.9307, 0.9363, 0.9284, 0.9198, 0.9146, 0.9126],
              [0.8204, 0.8951, 0.9303, 0.9364, 0.9282, 0.9197, 0.9144, 0.9126]])
# ax2.plot(x,y, '--')
ax2.set_xticks(np.arange(-3,5))
ax2.scatter(x, y.mean(axis=0), marker='o', s=20)
ax2.errorbar(x, y.mean(axis=0), yerr=3*y.std(axis=0), linestyle='--', capsize=4)
ax2.set_xlabel('Step Offset')
ax2.set_ylabel('Accuracy')
fig.tight_layout()
fig.savefig('plots_new/3.png', dpi=400)
```
