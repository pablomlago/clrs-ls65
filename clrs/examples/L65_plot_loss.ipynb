```python
import numpy as np
import re
import matplotlib.pyplot as plt
```

```python
file_path = "/rds/user/ar2217/hpc-work/L65/clrs-ls65/slurm-44998187.out"
with open(file_path, 'r') as file:
    lines = file.readlines()

pattern = re.compile(r'current loss ([\d\.]+)')
processor_type_pattern = re.compile(r'processor_type (\S+)')
regularisation_weight_pattern = re.compile(r'regularisation_weight ([\d\.]+)')

loss_values = []

for line in lines:
    match = pattern.search(line)
    if match:
        loss_value = float(match.group(1))
        loss_values.append(loss_value)

    # Search for 'processor_type'
    processor_type_match = processor_type_pattern.search(line)
    if processor_type_match:
        processor_type = str(processor_type_match.group(1))

    # Search for 'regularisation_weight'
    regularisation_weight_match = regularisation_weight_pattern.search(line)
    if regularisation_weight_match:
        regularisation_weight = float(regularisation_weight_match.group(1))

print(f"Processor Type: {processor_type}")
print(f"Regularisation Weight: {regularisation_weight}")

plt.semilogy(loss_values)
plt.xlabel('Iteration')
plt.ylabel('Loss')
```
